### Heroes Of Pymoli Data Analysis
* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_path = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_path)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
number_of_players=len(purchase_data["SN"].unique())
print(number_of_players)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
different_items=(len(purchase_data["Item ID"].unique()))
print(f"Unique Item Count: {different_items}")

average_price=purchase_data["Price"].mean()
print(f"Average Price: {average_price}")

purchase_count=len(purchase_data["Purchase ID"])
print(f"Number of Purchases: {purchase_count}")

revenue=purchase_data['Price'].sum()
print(f"Total Revenue: {revenue}")

purchase_analysis_df= pd.DataFrame(
                        {"Unique Item Count": [different_items],
                        "Average Price": [average_price],
                        "Number of Purchases": [purchase_count],
                        "Total Revenue": [revenue]})
purchase_analysis_df.head()
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
dropped_dups_df=purchase_data.drop_duplicates(subset="SN", keep="first")
dropped_dups_df["Gender"].value_counts()
gender_analysis = dropped_dups_df["Gender"].value_counts().rename_axis('Gender').reset_index(name='Counts')
gender_analysis["Percent"]=(gender_analysis["Counts"]/(len(dropped_dups_df))) * 100
gender_analysis=gender_analysis.round(2)
gender_analysis.head()
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_purchase_df=purchase_data.groupby(["Gender"])
total_purchase_count=gender_purchase_df["Purchase ID"].count()
avg_purchase=(gender_purchase_df["Price"].mean()).round(2)
total_value=gender_purchase_df["Price"].sum()
per_person_avg=(total_value/gender_purchase_df["Gender"].count()).round(2)

gender_df=pd.DataFrame({"Purchase Count": total_purchase_count,
                       "Average Purchase Price": avg_purchase,
                       "Total Purchase Value": total_value,
                       "Average Purchase Total by Person": per_person_avg})

gender_df.head()
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_bins=[0, 9, 14, 19, 24, 29, 34, 39, 150]
age_labels=["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

dropped_dups_df["Age Ranges"] = pd.cut(dropped_dups_df["Age"], age_bins, labels=age_labels, include_lowest=True)
age_range_df=pd.DataFrame(dropped_dups_df["Age Ranges"].value_counts())
age_range_df["Percentage of Players"]=((age_range_df["Age Ranges"]/(age_range_df["Age Ranges"].sum())) * 100).round(2)
age_range_df.sort_index()
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_data["Age Ranges"] = pd.cut(purchase_data["Age"], age_bins, labels=age_labels, include_lowest=True)
pdata_agebins=purchase_data.groupby("Age Ranges")
pdata_agebins.head()
```

```python
age_t_count=pdata_agebins["Purchase ID"].count()
age_avg_price=(pdata_agebins["Price"].mean()).round(2)
age_t_value=pdata_agebins["Price"].sum()
age_personavg=(age_t_value/age_range_df["Age Ranges"]).round(2)

age_panalysis_df=pd.DataFrame({"Purchase Count": age_t_count,
                              "Average Purchase Price": age_avg_price,
                              "Total Purchase Value": age_t_value,
                              "Average Total Purchase Per Person": age_personavg})
age_panalysis_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
purchase_data.head()
```

```python
sn_groups=purchase_data.groupby(["SN"])
sn_groups_df=pd.DataFrame(sn_groups["Price"].sum())


sn_groups_df["Purchase Count"]=(sn_groups["Purchase ID"].count())
sn_groups_df["Average Purchase Price"]=(sn_groups_df["Price"]/sn_groups_df["Purchase Count"]).round(2)
sn_groups_df=sn_groups_df.rename(columns={"Price":"Total Price"})


top_spenders_df=sn_groups_df.sort_values("Total Price", ascending=False)
top_spenders_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
item_df=purchase_data[["Item ID", "Item Name", "Price"]]
item_groups=item_df.groupby(["Item ID","Item Name"])

item_pcount=item_groups["Item ID"].count()
item_pshow=(item_groups["Price"].sum()/item_pcount).round(2)
item_pvalue=item_groups["Price"].sum()

top_items_df=pd.DataFrame({"Purchase Count": item_pcount,
                          "Price": item_pshow,
                          "Total Purchase Value": item_pvalue})
top_items_df.sort_values("Purchase Count", ascending= False)
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
sorted_item_df=top_items_df.sort_values("Total Purchase Value", ascending=False)
sorted_item_df.head()
```
